```
import pandas as pd
import numpy as np
df=pd.DataFrame(columns=['x','y'])
df['y']=np.random.randn(10)
df['x']=5
print(df)
print('\n')
print(df.mean())
print(df.std())
print('\n')

from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
scaler.fit(df[['y']])

print('\n')
print(scaler.mean_)
print('\n')

df['z']=scaler.transform(df[['y']])
print(df)
print('\n')
print(df['y'].to_numpy())
print('\n')
print(df['z'].to_numpy())
```
