```
import time
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
from sklearn.neural_network import MLPClassifier

# Load data from https://www.openml.org/d/554
X, y = fetch_openml('mnist_784', version=1, return_X_y=True)

#scale the pixel values to 0-1 range
X = X / 255

# use the traditional train/test split
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = y[:60000], y[60000:]

time.sleep(1) #Q2
mlp = MLPClassifier(hidden_layer_sizes=(50,), max_iter=10, alpha=1e-4,
                    solver='sgd', verbose=10, learning_rate_init=.1)


mlp.fit(X_train, y_train)

print("\nCPU Time: ",time.process_time())#Q2

print("\nTraining set score: %f" % mlp.score(X_train, y_train))
print("Test set score: %f" % mlp.score(X_test, y_test))

```

```
import time
from sklearn.tree import DecisionTreeClassifier
time.sleep(1)
decision_tree = DecisionTreeClassifier(max_depth=5)
decision_tree.fit(X_train, y_train)

x1 = time.process_time()
print("\nCPU Time: ",x1)#Q2
print("Training set score: %f" % decision_tree.score(X_train, y_train))
print("Test set score: %f" % decision_tree.score(X_test, y_test))
```
